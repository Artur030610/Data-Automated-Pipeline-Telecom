```python
import pandas as pd
import numpy as np

ruta = r"C:\Users\josperez\Documents\A-DataStack\01-Proyectos\01-Data_PipelinesFibex\02_Data_Lake\raw_data\1-Recaudación\3-Pago virtual\Data-Recaudacion Virtual 01-04-2026 al 14-04-2026.xlsx"
df = pd.read_excel(ruta, engine="calamine")

display(df.describe(include="all"))
display(df.dtypes)
display(df.columns)
display(df['Usuario'].value_counts())
df['Categoria'] = np.where(df['Usuario'] == 'APICONEXVE', 'App', 'Manual')
cols = ['N° Abonado', 'Fecha','Forma de Pago','Usuario', 'Oficina Cobro', 'Categoria']
df = df[cols]
display(df.info())

df_group = df.groupby(['Usuario', 'Fecha', 'Forma de Pago','Oficina Cobro']).agg(cantidad=('N° Abonado', 'size')).reset_index()
display(df_group)
mask_oficina = df['Oficina Cobro'].str.contains('VIRTUAL')
mask_usuario = df['Usuario'] == 'APICONEXVE'

display(df[mask_usuario].groupby(['Oficina Cobro', 'Usuario', 'Forma de Pago']).agg(cantidad=('N° Abonado', 'size')).reset_index())
display(df[mask_oficina].groupby(['Oficina Cobro', 'Usuario', 'Forma de Pago']).agg(cantidad=('N° Abonado', 'size')).reset_index())
display(df[mask_oficina].groupby(['Categoria']).agg(cantidad=('N° Abonado', 'size')).reset_index())
```
