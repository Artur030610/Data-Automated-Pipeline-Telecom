```python
import pandas as pd

# 1. Carga de datos
df_ONT = pd.read_parquet(r'C:\Users\josperez\Documents\A-DataStack\01-Proyectos\01-Data_PipelinesFibex\02_Data_Lake\gold_data\ONT_OFF_Gold.parquet')
df_total = pd.read_excel(r'C:\Users\josperez\Downloads\ACTIVOS POWER FAIL.xlsx', sheet_name='SMART POWER FAIL')

# --- SOLUCIÓN AL ERROR DE TIPO ---
# Convertimos ambas columnas de unión a String y quitamos espacios en blanco
df_ONT['N° Abonado'] = df_ONT['N° Abonado'].astype(str).str.strip()
df_total['Name'] = df_total['Name'].astype(str).str.strip()
display(df_total)
display(df_total.columns)
df_total = df_total[df_total['Estatus'] == 'ACTIVO'] # Filtramos solo los activos
# 2. Merge tipo 'outer' con indicador
df_merge = pd.merge(
    df_ONT, 
    df_total, 
    how='outer', 
    left_on='N° Abonado', 
    right_on='Name', 
    indicator=True
)

# 3. Filtrar para mantener solo lo que está en la derecha (los que no están en el Gold)
df_solo_derecha = df_merge[df_merge['_merge'] == 'right_only'].copy()

# 4. Limpieza
df_solo_derecha = df_solo_derecha.drop(columns=['_merge'])

display(df_solo_derecha)
```
